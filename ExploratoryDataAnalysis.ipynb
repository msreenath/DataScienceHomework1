### 3) Linear Regression model for GDP per capita
GDP per capita is the value of all final goods and services produced within a nation in a given year, converted at market exchange rates to current U.S. dollars, divided by the average (or mid-year) population for the same year. -Wikipedia
In our dataset, GDP per capita can be defined as ("GDP (PPP) in US $"/"Population") of the country

So, we will look at factors in close correlation with GDP Per Capita
- From correlation plots, it is clear that literacy rate and life expectancy are the parameters in close correlation with GDP Per Capita.
- So, we will try to create a single-variate linear regression model using life expectancy and then a multi-variate linear regression model using both parameters

#### Linear regression model of GDP per capita using life expectancy

```python
sns.lmplot("Life Expectancy", "GDP per capita", countryFactBook);
```

- Now, we shall compute the parameters of this linear regression model

```python
import pylab as pl
import numpy as np
import pandas as pd
import statsmodels.api as sm

dataset = countryFactBook.ix[:, ['GDP per capita', 'Life Expectancy', 'Country Name']].dropna()

gdppercapita = dataset['GDP per capita']
lifeexpectancy = dataset['Life Expectancy']

y = np.matrix(gdppercapita).transpose()
x1 = np.matrix(lifeexpectancy).transpose()
x = np.column_stack([x1])

X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()

print 'Coefficients: ', f.params[0:1]
print 'Intercept: ', f.params[1]
print 'P-Values: ', f.pvalues
print 'R-Squared: ', f.rsquared
```

-  GDP Per Capita = a0 + a1(Life Expectancy), where a0 is the intercept in above equation and a1 is the Coeffecient, which resulted in a R-squared value of 0.33, which is we will try to improve using the multi-variate linear regression model.

#### We use Multivariate Linear Regression to model "GDP Per Capita" variance with "Life Expectancy" and "Literacy rate (%)"

```python
dataset = countryFactBook.ix[:, ['GDP per capita', 'Literacy Rate (%)', 'Life Expectancy']].dropna()

gdppercapita = dataset['GDP per capita']
lendingrate = dataset['Literacy Rate (%)']
lifeexpectancy = dataset['Life Expectancy']

y = np.matrix(gdppercapita).transpose()
x1 = np.matrix(lendingrate).transpose()
x2 = np.matrix(lifeexpectancy).transpose()
x = np.column_stack([x1,x2])

# create linear model and fit it to data
X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()

print 'Coefficients: ', f.params[0:2]
print 'Intercept: ', f.params[2]
print 'P-Values: ', f.pvalues
print 'R-Squared: ', f.rsquared
```

-  GDP Per Capita = a0 + a1(Literacy rate) + a2(Life expectancy), where a0 is the intercept, a1 is the first coeffecient value and  a2 is second coeffecient value. This resulted in a R-squared value of 0.32, which is lesser than previous regression formulae.

##### Now we shall do another attempt at social welfare metrix considering all parameters

###### Second attempt
Steps followed
- Normalize the parameters with z-scores and compute the social welfare coeffecient
- Rank countries on basis of this and get at a conclusion whether it is good

```python
import scipy
normalizeddata = countryFactBook.ix[:,['Literacy Rate (%)','Health Expenditure/GDP','Net Migration Rate/1000 Population','Life Expectancy','Country Name']].dropna()
normalizeddata['Literacy Rate (%)'] = scipy.stats.zscore(normalizeddata['Literacy Rate (%)'])
normalizeddata['Health Expenditure/GDP'] = scipy.stats.zscore(normalizeddata['Health Expenditure/GDP'])
normalizeddata['Net Migration Rate/1000 Population'] = scipy.stats.zscore(normalizeddata['Net Migration Rate/1000 Population'])
normalizeddata['Life Expectancy'] = scipy.stats.zscore(normalizeddata['Life Expectancy'])
```

```python
#Now computing social welfare (we assume each parameter contributes to the equal amount, hence the value 0.25)
normalizeddata['social_welfare_score']= (normalizeddata['Literacy Rate (%)'] * 0.25) + (normalizeddata['Health Expenditure/GDP'] * 0.25) +  (normalizeddata['Life Expectancy'] * 0.25) +(normalizeddata['Net Migration Rate/1000 Population'] * -0.25)
```

###### The list of all countries in decreasing order of social welfare (best countries in terms of social welfare at first)

```python
normalizeddata = normalizeddata.sort(columns=['social_welfare_score'], axis=0, ascending=False)
printdata  = normalizeddata[['Country Name','social_welfare_score']]
printfullinline(printdata)
```

###5) Distance measure of how similar/different pairs of countries are

- Distance measurement can be done by taking the value of a country in a column and finding the cosine difference of the value with values of same column for all other countries

- This can be repeated for one or more than 1 column (i.e. all columns that we consider contributes to the difference)

- Since we have already computed a social welfare function (that showed expected outputs) which took many parameters into consideration, we will use the social welfare function along with few other paramters (internet users and population to further similarize the countries). 

####Other parameters

- Internet users gives an estimate of how developed the country is.

- Population gives an estimate of the country's size. We are not considering density of population here.

- ((Population - Labor force) / (Population)) to measure the unemplotment in the country.

- We are not considering GDP per capita. (Example)Consider GDP per capita of India and US. Though there is a big difference, we cannot expect that the standard of living in the 2 countries varies as much as the difference in GDP pre capita, as the currency conversion takes a toll (4000$ GDP per capita of India is a considerable amount in India in Rs, whereas it is very less in US)

- Though we cannot define a single metric (that fits all) for getting results in comparison of countries, we will take above paramters into consideration and define an approximate metric. So, we will assume equal weights to all the 4 parameters and arrive at a comparison of countries.

So, 4 parameters we are considering are:
- Social welfare
- Internet Users
- Population
- Unemployment

Formation of the metric to be considered for the calculation

```python
#Using dropna() function to drop rows with null values as it causes only 10 elements to be removed froom our sample.
#It is better  this way than arriving at wrong conclusions
normalizeddata = countryFactBook.ix[:,['Population','Labor Force','Internet Users (circa 2009)','Literacy Rate (%)','Health Expenditure/GDP','Net Migration Rate/1000 Population','Life Expectancy','Country Name']].dropna()
normalizeddata['Literacy Rate (%)'] = scipy.stats.zscore(normalizeddata['Literacy Rate (%)'])
normalizeddata['Health Expenditure/GDP'] = scipy.stats.zscore(normalizeddata['Health Expenditure/GDP'])
normalizeddata['Net Migration Rate/1000 Population'] = scipy.stats.zscore(normalizeddata['Net Migration Rate/1000 Population'])
normalizeddata['Internet Users (circa 2009)'] = scipy.stats.zscore(normalizeddata['Internet Users (circa 2009)'])
normalizeddata['Life Expectancy'] = scipy.stats.zscore(normalizeddata['Life Expectancy'])
```

```python
normalizeddata['Unemployment'] = np.divide(np.subtract(normalizeddata['Population'], normalizeddata['Labor Force']),normalizeddata['Population'])
normalizeddata['Unemployment'] = scipy.stats.zscore(normalizeddata['Unemployment'])
normalizeddata['Population'] = scipy.stats.zscore(normalizeddata['Population'])
normalizeddata['social_welfare_score']= (normalizeddata['Literacy Rate (%)'] * 0.25) + (normalizeddata['Health Expenditure/GDP'] * 0.25) +  (normalizeddata['Life Expectancy'] * 0.25) +(normalizeddata['Net Migration Rate/1000 Population'] * -0.25)
normalizeddata['distancemetric'] = (normalizeddata['social_welfare_score'] * 0.25) +  (normalizeddata['Population'] * 0.25) +(normalizeddata['Internet Users (circa 2009)'] * 0.25 + (normalizeddata['Unemployment'] * 0.25))
table1 = normalizeddata.copy()
table2 = normalizeddata.copy()
```

Comparison of distance metric column of one country across all others and finding the maximum and minimum variation

```python
# Taking outer join to get table of one country to all
table1['joincolumn'] = 0
table2['joincolumn'] = 0
dataset = pd.merge(table1, table2, on='joincolumn', how='outer')
dataset['distancemetric'] = abs(dataset['distancemetric_x'].astype(float) - dataset['distancemetric_y'].astype(float))
# Remove same country as that being compared from the list
dataset = dataset[dataset['distancemetric'] != 0]
finaldata = dataset.groupby('Country Name_x')['distancemetric']
variationoncountry = pd.DataFrame(columns=['minimumvariation','maximumvariation'])
# Find the minimum and maximum variation of the distance metric
variationoncountry['minimumvariation'] = finaldata.idxmin()
variationoncountry['maximumvariation'] = finaldata.idxmax()
```

####Closest pairs

```python
nearestneighbour = pd.merge(variationoncountry, dataset, left_on = 'minimumvariation', suffixes=('_x', '_y'), right_index=True)
printfullinline(nearestneighbour['Country Name_y'])
```

####Farthest pairs

```python
farthestneighbour = pd.merge(variationoncountry, dataset, left_on = 'maximumvariation', suffixes=('_x', '_y'), right_index=True)
printfullinline(farthestneighbour['Country Name_y'])
```
