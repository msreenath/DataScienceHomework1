### 3. Linear Regression model for GDP per capita
GDP per capita is the value of all final goods and services produced within a nation in a given year, converted at market exchange rates to current U.S. dollars, divided by the average (or mid-year) population for the same year. -Wikipedia
In our dataset, GDP per capita can be defined as ("GDP (PPP) in US $"/"Population") of the country

So, we will look at factors in close correlation with GDP Per Capita
- From correlation plots, it is clear that literacy rate and life expectancy are the parameters in close correlation with GDP Per Capita.
- So, we will try to create a multi-variate linear regression model using these parameters

#### We use Multivariate Linear Regression to model "GDP Per Capita" variance with "Life Expectancy" and "Commercial Prime Lending Rate"

```python
import pylab as pl
import numpy as np
import pandas as pd
import statsmodels.api as sm

dataset = countryFactBook.ix[:, ['GDP per capita', 'Commercial Prime Lending Rate', 'Life Expectancy']].dropna()

gdppercapita = dataset['GDP per capita']
lendingrate = dataset['Commercial Prime Lending Rate']
lifeexpectancy = dataset['Life Expectancy']

y = np.matrix(gdppercapita).transpose()
x1 = np.matrix(lendingrate).transpose()
x2 = np.matrix(lifeexpectancy).transpose()
x = np.column_stack([x1,x2])

# create linear model and fit it to data
X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()

print 'Coefficients: ', f.params[0:2]
print 'Intercept: ', f.params[2]
print 'P-Values: ', f.pvalues
print 'R-Squared: ', f.rsquared
```

####Linear regression expression

-  GDP Per Capita = a0 + a1(Life Expectancy) + a2(Commercial Prime Lending Rate), where a0 = 808.48, a1 = -33681.28 and a2 = -758.16, which resulted in a R-squared value of 0.48, which is the closest we could get to.

```python
countryFactBook['GDP per capita (calc)'] = np.subtract(np.subtract(808.48,(np.multiply(33681,countryFactBook['Commercial Prime Lending Rate']))), (np.multiply(33681,countryFactBook['Life Expectancy'])))
```

```python
plt.scatter(lendingrate, gdppercapita)  # Plot the raw data
plt.xlabel("Gross Domestic Product")
plt.ylabel("Lending rate and life expectancy")
Y = 808.48 - (33681 * lendingrate) - (758.16 * lifeexpectancy)
plt.plot(lendingrate, Y, 'r')
```

### 4. Ranking function to measure general social welfare

####Description of the measure used
####Parameters that affect general social welfare
-  Health Expenditure
-  Literacy rate
-  Life expectancy

####Parameters that reflect general social welfare
-  Net migration rate / 1000

Since there can be many parameters, other than those in our data, affecting general social welfare. 
We will resort to net migration rate / 1000 as a measure of general social welfare, as this can be mapped as a direct consequence of general social welfare (in a broad sense)

```python
netmigration= countryFactBook.ix[:,['Country Name','Net Migration Rate/1000 Population']].dropna()
rankedmigration = netmigration.sort(columns=['Net Migration Rate/1000 Population'] , axis=0, ascending=True)
```
